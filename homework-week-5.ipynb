````python
url = "https://apps.health.ny.gov/pubdoh/professionals/doctors/conduct/factions/AlphabetSearchAction.action?alpbhabetSearch=Z&d-49653-p={}"
````

````python
total_pages = 6 
mylinks = [url.format(number) for number in range (1, total_pages)]
mylinks
````

````python
from random import randrange
import time
````

````python
all_df = []
busted_links = []

counter = 1
for link in mylinks:
    print(f"Scraping {counter} of {len(mylinks)}")
    counter += 1
    response = requests.get(link)
    if response.status_code == 200:
        df = pd.read_html(response.text)
        all_df.append(df[0])
    else:
        print(f"{link} returned a busted link with {response.code_status}")
    snoozer = randrange(5,12)
    print(f"snoozing for {snoozer} seconds before next scrape")
    time.sleep(snoozer)
    
print("All done....")
````

````python
all_df
````

````python
df = pd.concat(all_df, ignore_index = True)
df
````
